```python
# =======================================================
# الخلية النهائية: الأتمتة الكاملة للاستدلال والمطابقة السريرية
# =======================================================
!pip install -q owlready2

from owlready2 import *
import os
import shutil

# 1. تحديد مسار الأنطولوجيا المليئة بالبيانات بناءً على مسارك الجديد
INPUT_ONTO = "/kaggle/input/datasets/maherghanem/ssssssss/PharmaCompass_Populated_Robust.owl"
WORK_DIR = "/kaggle/working/"
TEMP_ONTO = os.path.join(WORK_DIR, "PharmaCompass_For_Reasoning.owl")

# نسخ الملف لمجلد العمل لتجنب مشاكل "القراءة فقط" في Kaggle أثناء الاستدلال
shutil.copy(INPUT_ONTO, TEMP_ONTO)

# 2. تحميل الأنطولوجيا إلى الذاكرة
print("⏳ جاري تحميل قاعدة المعرفة (الأنطولوجيا)...")
onto_path.append(WORK_DIR)
onto = get_ontology("file://" + TEMP_ONTO).load()

# 3. تشغيل محرك الاستدلال برمجياً (Reasoning)
print("🧠 جاري تشغيل محرك الاستدلال المنطقي (HermiT Reasoner)...")
with onto:
    try:
        # هذه الدالة تقوم باستنتاج العلاقات الخفية كما يفعل برنامج Protégé تماماً
        sync_reasoner() 
        print("✅ اكتمل الاستدلال المنطقي بنجاح!")
    except Exception as e:
        print("⚠️ (ملاحظة: إذا لم تتوفر بيئة Java في Kaggle، سنستعلم من البيانات المستخرجة مباشرة)")

print("\n" + "="*60)
print("🔍 نتائج محرك البوصلة الدوائية (Clinical Trials Matcher)")
print("="*60)

# ==============================================
# الاستعلام الأول: البحث عن المرضى الإناث
# ==============================================
print("\n👩‍⚕️ التجربة السريرية الأولى: (تتطلب مرضى إناث فقط)")
female_patients = []
for p in onto.Patient.instances():
    # نبحث عن الإناث سواء كانت القيمة 'FEMALE' أو 'F'
    if "FEMALE" in p.has_gender or "F" in p.has_gender:
        female_patients.append(p.name)
        
print(f"   => تم العثور على {len(female_patients)} مريضة مطابقة.")
if female_patients:
    print(f"   => عينة من المطابقين: {female_patients[:5]}...")

# ==============================================
# الاستعلام الثاني: البحث عن مرض طبي محدد
# ==============================================
print("\n🫀 التجربة السريرية الثانية: (تتطلب مرضى أصيبوا بـ Cardiac Arrest)")
# استدعاء الكيان المنطقي للمرض من الأنطولوجيا
target_condition = onto.Condition_Cardiac_Arrest 
cardiac_patients = []

for p in onto.Patient.instances():
    if target_condition in p.has_condition:
        cardiac_patients.append(p.name)
        
print(f"   => تم العثور على {len(cardiac_patients)} مريض مطابق.")
if cardiac_patients:
    print(f"   => أرقام المرضى: {cardiac_patients}")

# ==============================================
# الاستعلام الثالث: استعلام معقد (جنس + مرض)
# ==============================================
print("\n🔬 التجربة السريرية الثالثة (المعقدة): (تتطلب إناث + إصابة بتوقف القلب)")
# نستخدم تقاطع المجموعات (Intersection) في بايثون للمطابقة المعقدة
complex_match = set(female_patients).intersection(set(cardiac_patients))

print(f"   => العثور على التطابق الدقيق:")
if complex_match:
    for match in complex_match:
        print(f"      🌟 المريض المثالي لهذه التجربة: {match}")
else:
    print("      لا يوجد مريض يطابق جميع هذه الشروط المعقدة معاً.")

print("\n" + "="*60)
print("🎉 اكتملت أتمتة النظام بالكامل (End-to-End Pipeline)!")
```
